```python
import tensorflow as tf
import numpy as np
import utils

dir_saved_model = 'saved_models'
model_id = '20180815180530'
model_to_import = '{}/{}-fnn-sigm-24-adam-lr0.001000-bs32-epoch1376-testacc1.000.ckpt'.format(model_id, model_id)

saver = tf.train.import_meta_graph('{}/{}.meta'.format(dir_saved_model, model_to_import))
#saver = tf.train.Saver()

config = tf.ConfigProto()
config.gpu_options.allow_growth = True
```

```python
print([node.name for node in tf.get_default_graph().as_graph_def().node])
```

```python
tf.trainable_variables()
```

```python
decimal_n1 = 255 # range: [0,255]
decimal_n2 = 2   # range: [0,255]
n_binary_digits = 8

input_test = utils.get_fcn_input(decimal_n1, decimal_n2, n_binary_digits)
target_test = utils.get_fcn_target(decimal_n1, decimal_n2, n_binary_digits)

with tf.Session(config=config) as sess:
    saver.restore(sess, '{}/{}'.format(dir_saved_model, model_to_import))
    print("Model restored.")
    
    # End of all epochs 
    # Run computing test loss, accuracy
    [predictions_value] = sess.run(
        ['Cast:0'],
        feed_dict={'inputs:0':input_test, 'outputs:0':target_test})
    
    print("{} + {} = {}".format(decimal_n1, decimal_n2, utils.decode_fcn_output(predictions_value)))
```

```python
with tf.Session(config=config) as sess:
    saver.restore(sess, '{}/{}'.format(dir_saved_model, model_to_import))
    print("Model restored.")
        
    n_binary_digits = 8
    all_correct = True
    for i in range(2**n_binary_digits):
        print("{}iterations".format(i * 2**n_binary_digits))
        for j in range(2**n_binary_digits):
            decimal_n1 = i  # range: [0,255]
            decimal_n2 = j  # range: [0,255]

            input_test = utils.get_fcn_input(decimal_n1, decimal_n2, n_binary_digits)
            target_test = utils.get_fcn_target(decimal_n1, decimal_n2, n_binary_digits)

            [predictions_value] = sess.run(
                ['Cast:0'],
                feed_dict={'inputs:0':input_test, 'outputs:0':target_test})
            if decimal_n1 + decimal_n2 != utils.decode_fcn_output(predictions_value):
                all_correct = False
                break
        if all_correct == False:
            break
    print("All corrected? {}".format(all_correct))
```
